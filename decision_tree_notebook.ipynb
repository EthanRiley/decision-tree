```python
import pandas as pd
import numpy as np
from collections import Counter
import math
from decision_tree import *
```

```python
training_data = pd.read_csv('spotify_train.csv')
training_data.head()
```

```python
training_data.describe()
```

```python
Counter(training_data['track_genre']).most_common(1)[0][1]
```

```python
tree = dtree(training_data, gini, class_col='track_genre')
```

```python
from pprint import pprint
pprint(tree, width=25)
```

```python
testing_data = pd.read_csv('spotify_test.csv')
```

```python
predictions = predict(tree, testing_data)
predictions
```

```python
# Compare predictions array to the actual values in the testing_data dataframe to determine accuracy

num_correct = sum([1 if predictions[i] == testing_data['track_genre'][i] else 0 for i in range(len(predictions))])

total = len(predictions)

accuracy = num_correct / total
```
